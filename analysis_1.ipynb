```python
# x-axis = n, y-xis=wait_time
import pandas as pd
df = pd.read_csv('reports_fcfs.csv')
df.head()
```

```python
p1 = np.array(df.iloc[np.where(df['process']==1)].sort_values(by=['n']))
p2 = np.array(df.iloc[np.where(df['process']==2)].sort_values(by=['n']))
p3 = np.array(df.iloc[np.where(df['process']==3)].sort_values(by=['n']))
```

```python
# p1[:,2]
```

```python
fig, ax = plt.subplots()
fig.set_figheight(10)
fig.set_figwidth(20)
ax.plot(p1[3:63, 1], p1[3:63, 3], "r", label="Process 1")
ax.plot(p2[3:63, 1], p2[3:63, 3], "g", label = "Process 2")
ax.plot(p3[3:63, 1], p3[3:63, 3], "b", label = "Process 3")
plt.title("Turn around time VS n")
plt.xlabel("n")
plt.ylabel("TAT (micro seconds)")
leg = ax.legend()
plt.show()
```

## Comparision between FCFS and Round Robin

```python
fcfs.head()
```

```python
rr.head()
```

```python
p1 = np.array(fcfs.iloc[np.where(fcfs['process']==1)].sort_values(by=['n']))
p2 = np.array(fcfs.iloc[np.where(fcfs['process']==2)].sort_values(by=['n']))
p3 = np.array(fcfs.iloc[np.where(fcfs['process']==3)].sort_values(by=['n']))
```

```python
t1 = rr.iloc[np.where(rr['tq']==2)]
t2 = rr.iloc[np.where(rr['tq']==5)]
t3 = rr.iloc[np.where(rr['tq']==8)]
t1_p1 = np.array(t1.iloc[np.where(t1['process']==1)].sort_values(by=['n']))
t1_p2 = np.array(t1.iloc[np.where(t1['process']==2)].sort_values(by=['n']))
t1_p3 = np.array(t1.iloc[np.where(t1['process']==3)].sort_values(by=['n']))
t2_p1 = np.array(t2.iloc[np.where(t2['process']==1)].sort_values(by=['n']))
t2_p2 = np.array(t2.iloc[np.where(t2['process']==2)].sort_values(by=['n']))
t2_p3 = np.array(t2.iloc[np.where(t2['process']==3)].sort_values(by=['n']))
t3_p1 = np.array(t3.iloc[np.where(t3['process']==1)].sort_values(by=['n']))
t3_p2 = np.array(t3.iloc[np.where(t3['process']==2)].sort_values(by=['n']))
t3_p3 = np.array(t3.iloc[np.where(t3['process']==3)].sort_values(by=['n']))
```

```python
avg_fcfs = (p1[:,3]+p2[:,3]+p3[:,3])/3
```

```python
avg_rr_t1 = (t1_p1[:,3]+t1_p2[:,3]+t1_p3[:,3])/3
avg_rr_t2 = (t2_p1[:,3]+t2_p2[:,3]+t2_p3[:,3])/3
avg_rr_t3 = (t3_p1[:,3]+t3_p2[:,3]+t3_p3[:,3])/3
```

```python
# plt.plot(p1[:,1], avg_fcfs)
# plt.title('Average FCFS TAT vs n')
# plt.x
# plt.show()

fig, ax = plt.subplots()
fig.set_figheight(10)
fig.set_figwidth(20)
ax.plot(p1[:,1], avg_fcfs, "r", label="FCFS")
ax.plot(t1_p1[:,1], avg_rr_t1, "g", label = "RR T1")
ax.plot(t2_p2[:, 1], avg_rr_t2, "b", label = "RR T2")
ax.plot(t3_p2[:, 1], avg_rr_t3, "", label = "RR T3")
plt.title("Comparison between average TAT of RR and FCFS")
plt.xlabel("n")
plt.ylabel("TAT (micro seconds)")
leg = ax.legend()
plt.show()
```

```python
avg_fcfs = (p1[:,4]+p2[:,4]+p3[:,4])/3
avg_rr_t1 = (t1_p1[:,4]+t1_p2[:,4]+t1_p3[:,4])/3
avg_rr_t2 = (t2_p1[:,4]+t2_p2[:,4]+t2_p3[:,4])/3
avg_rr_t3 = (t3_p1[:,4]+t3_p2[:,4]+t3_p3[:,4])/3

fig, ax = plt.subplots()
fig.set_figheight(10)
fig.set_figwidth(20)
ax.plot(p1[:,1], avg_fcfs, "r", label="FCFS")
ax.plot(t1_p1[:,1], avg_rr_t1, "g", label = "RR T1")
ax.plot(t2_p2[:, 1], avg_rr_t2, "b", label = "RR T2")
ax.plot(t3_p2[:, 1], avg_rr_t3, "", label = "RR T3")
plt.title("Comparison between average WT of RR and FCFS")
plt.xlabel("n")
plt.ylabel("WT (micro seconds)")
leg = ax.legend()
plt.show()
```
